```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('/content/company_sales_data.csv')
df
```

```python
#1

profit=df['total_profit']
months=df['month_number']
plt.plot(months,profit)
```

```python
#2

plt.plot(months,profit,linewidth=3,marker='o',color='r',linestyle=':')
plt.legend(loc='lower right')
plt.xlabel('months Number')
plt.ylabel('sold unit number')
plt.title('company sales data')
```

```python
#3
toothpaste_count=df['toothpaste']
facewash=df['facewash']
facecream=df['facecream']
bathingsoap=df['bathingsoap']
shampoo=df['shampoo']
shampoo=df['shampoo']
moisturizer=df['moisturizer']
plt.plot(months,toothpaste_count)
plt.plot(months,facewash)
plt.plot(months,facecream)
plt.plot(months,bathingsoap)
plt.plot(months,shampoo)
plt.plot(months,moisturizer)
plt.xticks(months)
plt.yticks([1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000])
plt.xlabel('months')
plt.ylabel('sold unit number')
plt.title('company sales data')
plt.legend(['toothpaste','facewash','facecream','bathingsoap','shampoo'])
```

```python
#4
plt.scatter(toothpaste_count,months)
```

```python
#5
plt.bar([a-0.25 for a in months],facecream,width=0.25,label='Face cream sale', align='edge')
plt.bar([a+0.25 for a in months],facewash,width=-0.25,label="Face wash sale", align='edge')
plt.legend()
plt.xlabel('months')
plt.ylabel('sold unit number')
plt.title('company sales data')
plt.xticks(months)
plt.grid()
```

```python
plt.bar([a+0.40 for a in months],toothpaste_count,width=0.40,label='toothpaste sale', align='edge',color='r')
plt.bar([a-0.40 for a in months],bathingsoap,width=-0.40,label='bathingsoap sale', align='edge',color='g')
plt.legend()
plt.xlabel('months')
plt.ylabel('sold unit number')
plt.title('company sales data')
plt.xticks(months)
plt.grid()
```

```python
#7
lab=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(sales,labels=lab,autopct='%0.2f%%')
plt.title('company sales data')

plt.tight_layout()
```
